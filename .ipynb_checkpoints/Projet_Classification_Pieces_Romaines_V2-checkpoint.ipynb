# Le projet:

Ce projet se décompose en quatre étapes :


1.   Constitution de la base de données de plus de 30 000 photos (scraping). 
2.   Utilisation de [VOTT](https://github.com/microsoft/VoTT) afin d'étiqueter les images.

Les scripts qui ont permis la création de la base de données ne seront pas présents dans ce notebook.

# I. Constitution de la base de données:

La base de données est constituée de plus de 30 124 photos de pièces romaines représentant plusieurs empereurs. Cependant, cette base de données est très déséquilibrée. Celle-ci n'étant pas très volumineuse pour la quantité d'empereurs différents j'ai donc arbitrairement sélectionné quatre empereurs de cette base de données : CONSTANTIN Ier LE GRAND, GORDIEN III, PROBUS et POSTUME.
L'objectif sera donc de résoudre un problème de classification des pièces romaines ayant pour face le portrait de ces quatres empereurs. La base de données exploitable n'est, cependant, constituée que de 4454 photos. Malgré la taille réduite de la base de données, je n'ai pas eu besoin d'utiliser des techniques de data augmentation.

La consitution de la base de données s'est effectuée en une vingtaine d'heures.

## Classification :

L'échantillon train est constitué de 80% de la base de données soit de 3560 photos (0.8 * 4450).
L'échantillon test est constitué de 890 photos.

```python
def data_gen(target,copy_target):
    """ Générateur permettant de retourner une image et son label."""
    for i in range(3560):
        height = 200
        width = 200
        liste_empereurs = ['CONSTANTIN', 'GORDIEN',
       'PROBUS', 'POSTUME']
        try:
            train_image_fullpath = "E:/PROJETS/ROMAN COIN/CGB_crop/" + images[i]
            train_img = keras.preprocessing.image.load_img(train_image_fullpath, target_size=(height, width))
            train_img_arr = keras.preprocessing.image.img_to_array(train_img)
            one_hot = [0] * 4
            one_hot = np.array(one_hot)
            y_train = liste_empereurs.index(re.findall(r'[A-Za-z]+',images[i])[0])
            one_hot[y_train] = 1
            y_train = one_hot
            y_train = y_train.reshape(1,4)
            y_train = tf.convert_to_tensor(y_train)
            x_train = tf.expand_dims(train_img_arr, axis=0)
            yield (x_train,y_train)
        except:
            continue
        
```

```python
def data_gene_test(target,copy_target):
    """ Générateur permettant de retourner une image et son label."""
    for i in range(3560,len(images)):
        height = 200
        width = 200
        liste_empereurs = ['CONSTANTIN', 'GORDIEN',
       'PROBUS', 'POSTUME']
        try:
            test_image_fullpath = "E:/PROJETS/ROMAN COIN/CGB_crop/" + images[i]
            test_img = keras.preprocessing.image.load_img(test_image_fullpath, target_size=(height, width))
            test_img_arr = keras.preprocessing.image.img_to_array(test_img)
            one_hot = [0] * 4
            one_hot = np.array(one_hot)
            y_test = liste_empereurs.index(re.findall(r'[A-Za-z]+',images[i])[0])
            one_hot[y_test] = 1
            y_test = one_hot
            y_test = y_test.reshape(1,4)
            y_test = tf.convert_to_tensor(y_test)
            x_test = tf.expand_dims(test_img_arr, axis=0)
            yield (x_test,y_test)
        except:
            continue
```

```python
num_classes = 4

model = Sequential([
    layers.experimental.preprocessing.Rescaling(1./255, input_shape=(200, 200, 3)),
    layers.Conv2D(16, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(32, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(64, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Flatten(),
    layers.Dense(16, activation='relu'),
    layers.Dense(num_classes,activation="softmax")
])

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
    loss=tf.keras.losses.CategoricalCrossentropy(from_logits = False),
    metrics=['accuracy']
)
```

```python
model.summary()
```

```python
epochs = 10
acc = []
val_acc = []
loss = []
val_loss = []
for epoch in range(epochs):
    train_generator = data_gen(target,copy_target)
    test_generator = data_gene_test(target,copy_target)
    history = model.fit_generator(train_generator,steps_per_epoch=3560,validation_data=test_generator)
    acc.append(history.history['accuracy'])
    val_acc.append(history.history['val_accuracy'])
    loss.append(history.history['loss'])
    val_loss.append(history.history['val_loss'])
```

```python
epochs_range = range(epochs)

plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()  
```

```python
link_train = "E:/PROJETS/ROMAN COIN/DATA3/vott-csv-export/"
link_test = "E:/PROJETS/ROMAN COIN/TEST/"
label = pd.read_csv('ROMAN-Coin-export.csv')
```

```python
images_test = os.listdir(link_test)
dico_train = {"image" : [],"xmin" : [],"ymin":[],"xmax":[],"ymax":[],"label":[]}
dico_test = {"image" : [],"xmin" : [],"ymin":[],"xmax":[],"ymax":[],"label":[]}

for im in images_test:
    for i in range(len(label)):
        if label['image'].iloc[i] == im:
            dico_test['image'].append(label['image'].iloc[i])
            dico_test['xmin'].append(label['xmin'].iloc[i])
            dico_test['ymin'].append(label['ymin'].iloc[i])
            dico_test['xmax'].append(label['xmax'].iloc[i])
            dico_test['ymax'].append(label['ymax'].iloc[i])
            dico_test['label'].append(label['label'].iloc[i])
        elif label['image'].iloc[i] not in images_test:
            if label['image'].iloc[i] not in dico_train['image']:
                dico_train['image'].append(label['image'].iloc[i])
                dico_train['xmin'].append(label['xmin'].iloc[i])
                dico_train['ymin'].append(label['ymin'].iloc[i])
                dico_train['xmax'].append(label['xmax'].iloc[i])
                dico_train['ymax'].append(label['ymax'].iloc[i])
                dico_train['label'].append(label['label'].iloc[i])
label_test = pd.DataFrame(dico_test)
label_train = pd.DataFrame(dico_train)
```

```python
def coord_gen_train():
    """ Générateur permettant de retourner une image et les coordonnées de la bounding box associée."""
    for i in range(len(label_train)):
        height = 200
        width = 200
        liste_empereurs = ['CONSTANTIN', 'GORDIEN',
       'PROBUS', 'POSTUME']
        try:
            train_image_fullpath = link_train + label_train['image'].iloc[i]
            train_img = keras.preprocessing.image.load_img(train_image_fullpath, target_size=(height, width))
            train_img_arr = keras.preprocessing.image.img_to_array(train_img)
            one_hot = [0] * 4
            one_hot = np.array(one_hot)
            y_train = liste_empereurs.index(re.findall(r'[A-Za-z]+',label_train['image'].iloc[i])[0])
            one_hot[y_train] = 1
            y_train = one_hot
            y_train = y_train.reshape(1,4)
            y_train = tf.convert_to_tensor(y_train)
            x_train = tf.expand_dims(train_img_arr, axis=0)
            im = Image.open(train_image_fullpath)
            width2 , height2 = im.size
            w = width2 / 200
            h = height2 / 200
            xmin = round(label_train['xmin'].iloc[i]/ w, 2)/ 200
            ymin = round(label_train['ymin'].iloc[i]/ h, 2)/ 200
            xmax = round(label_train['xmax'].iloc[i]/ w, 2)/ 200
            ymax = round(label_train['ymax'].iloc[i]/ h, 2)/ 200
            yield (x_train,np.array([(xmin,ymin,xmax,ymax)]))
        except:
            continue
        
```

```python
def verif():
    """ Générateur permettant de retourner une image test et son label associé"""
    liste_empereurs = ['CONSTANTIN', 'GORDIEN',
       'PROBUS', 'POSTUME']
    list_pred = os.listdir("E:\PROJETS\ROMAN COIN\DATA2")
    random.shuffle(list_pred)
    for coin in list_pred:
        name = re.findall(r'[^ ]+',coin.split("-")[0])[0].upper()
        if name in liste_empereurs:
            train_img = keras.preprocessing.image.load_img("E:/PROJETS/ROMAN COIN/DATA2/"  + coin, target_size=(200, 200))
            train_img_arr = keras.preprocessing.image.img_to_array(train_img)                
            x_test = tf.expand_dims(train_img_arr, axis=0)
            pred = classification_model.predict_classes(x_test)
            pred_xy = bound.predict(x_test)
            xmin = pred_xy[0][0] * 200
            ymin = pred_xy[0][1]* 200
            xmax = pred_xy[0][2]* 200
            ymax = pred_xy[0][3]* 200
            im = Image.open("E:/PROJETS/ROMAN COIN/DATA2/"  + coin)
            width2 , height2 = im.size
            w = width2 / 200
            h = height2 / 200
            newsize = (200, 200)
            im1 = im
            plt.imshow(im1)
            ax = plt.gca()
            rect = patches.Rectangle((xmin*w,ymin*h),
                 xmax*w,
                 ymax*h,
                 linewidth=2,
                 edgecolor='red',
                 fill = False)
            ax.add_patch(rect)
            plt.show()
            yield (liste_empereurs[pred[0]],name)
```
